```python
def twoSum(nums,target):
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            if nums[i]+nums[j] ==target:
                return [i,j]
nums = [2,7,11,5]
target =9
twoSum(nums,target)
```

```python
def threeSum(alist):
    alist.sort()
    n = len(alist)
    res =[]
    for i in range(n):
        low = i+1
        high = n-1
        if i>0 and alist[i] ==alist[i-1]:continue
        while low<high:
            if alist[i]+alist[low]+alist[high] ==0:
                res.append([alist[i],alist[low],alist[high]])
                low+=1
            if alist[i]+alist[low]+alist[high]>0:
                high-=1
            if alist[i]+alist[low]+alist[high]<0:
                low+=1
    return res
nums = [-1, 0, 1, 2, -1, -4]
s=threeSum(nums)
```

```python
def countAndSay(n):
    if n ==1:return "1"
    if n==2: return "11"
    pre ="11"
    for i in range(3,n+1):
        res=""
        cnt=1
        
        length = len(pre)
        for j in range(1,length):
            if pre[j-1] ==pre[j]:
                cnt+=1
            else:
                res+=str(cnt)+pre[j-1]
        res+=str(cnt)+pre[j]
        pre = res
    return res
```

```python
def maxSlide(nums,k):
    n =len(nums)
    if n*k ==0:return []
    return [max(nums[i:i+k],default=0) for i in range(n-k+1)]
```

```python
nums = [1, 2, 3, 6, 8, 3, 6, 9, 10]
k = 4
maxSlide(nums,k)
```

```python
def lengthOF(nums):
    size = len(nums)
    if size<2:return size
    LIS = [nums[0]]
    for num in nums[1:]:
        if num>LIS[-1]:
            LIS.append(num)
            continue
        l,r = 0,len(LIS)-1
        while l<r:
            mid =1+(r-1)//2
            if LIS[mid]<num:
                l=mid+1
            else:
                r =mid
        LIS[l] =num
    return len(LIS)
    
```
