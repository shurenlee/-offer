```python

def myAtoi(str):
    str = str.lstrip()  # 
    if len(str) == 0:
        return 0
    # 
    index = 0
    # 1
    if str[0] in ["-","+"]:
        index = 1
    # 
    if len(str[index:]) == 0 or not str[index].isdecimal():
        return 0
    # 
    while index < len(str):
        if not str[index].isdecimal():
            break
        index += 1
    # 
    str = int(str[:index])
    # 
    if str < -2**31:
        return -2**31
    if str > 2**31 -1:
        return 2**31 -1
    return str
myAtoi('   -42')
```

### 50. Pow(x, n)

```python
def mypow(x,n):
    res = 1
    for i in range(n):
        res*=x
    return res
```

```python
mypow(2,3)
```

```python
def pow2(x,n):
    if n ==0:
        return 1
    if n<0:
        return 1/pow2(x,-n)
    if n%2 ==1:
        return x* pow2(x,n-1)
    return pow2(x*x,n/2)
```

```python
pow2(2,3)
```

```python
nums = [int(each) for each in input().split()]
def majority(alist):
    dic = {}
    for item in alist:
        dic[item]=dic.get(item,0)+1
    res=sorted(dic.items(),key = lambda s:s[1])
    return res[-1][0]
majority(nums)
```

```python
def trapp(height):
    res = []
    for i in range(1,len(height)):
        s =min(max(height[0:i+1],default=0),max(height[i+1:len(height)+1],default=0))-height[i]
        if s>0:
            res.append(s)
    return res
```

```python
height =[0,1,0,2,1,0,1,3,2,1,2,1]
trapp(height)
```
